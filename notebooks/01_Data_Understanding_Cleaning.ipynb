# Netflix Content Analysis – Data Understanding & Cleaning

This notebook is the **first phase** of the Netflix Content Analysis project.  
In this phase, we will:
- Import the Netflix dataset
- Inspect its structure, data types, and missing values
- Perform basic cleaning to prepare it for feature engineering and exploratory data analysis

**Dataset Source:** DeepDataLake – Netflix Movies & TV Shows Metadata (~8,000 rows, CC0 licensed)


```python
import pandas as pd
import numpy as np

df = pd.read_csv("data.csv")  # Replace with your actual filename
df.head()
```

## 2. Basic Dataset Information

We will now:
- Check the shape of the dataset (rows × columns)
- View the column names
- Get a concise summary of data types and non-null counts


```python

print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.\n")

print("Column Names:", df.columns.tolist(), "\n")

df.info()
```

## 3. Missing Values Check

To ensure data quality, let's identify:
- Which columns have missing values
- The percentage of missing values per column


```python

missing_counts = df.isnull().sum()
missing_percent = (missing_counts / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing_counts,
    'Percentage': missing_percent
}).sort_values(by='Percentage', ascending=False)

missing_df
```

## 4. Initial Observations & Data Cleaning Plan

From the missing values check, we observe:
- `director` has ~30% missing values — expected, as some shows list no specific director
- `country` and `cast` have ~9% missing values — we can keep them as NaN for now, as they may still be useful later
- `date_added`, `rating`, `duration` have very few missing values — we can handle them individually
- `show_id`, `type`, `title`, `release_year`, `listed_in`, and `description` have no missing values

**Cleaning Plan:**
1. Convert `date_added` to datetime format
2. Standardize text columns by stripping spaces
3. Decide whether to fill or leave missing values based on column importance
4. Ensure `release_year` and other numeric fields are in the correct format


```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed', errors='coerce')
df.head()
```

. Explore Key Categorical Columns

We will check:
- The unique values in `type`, `rating`, and `duration`
- This will help us understand possible inconsistencies before cleaning


```python
print("Unique values in 'type':", df['type'].unique(), "\n")
print("Unique values in 'rating':", df['rating'].unique(), "\n")
print("Unique values in 'duration':", df['duration'].unique(), "\n")
```


 Handling Missing Values

Our approach:
- `director`: Keep as NaN for now, as missing directors are common for shows and some content types
- `country` and `cast`: Keep as NaN for now; we may use them in specific analyses even if incomplete
- `date_added`, `rating`, `duration`: Fill missing values with suitable defaults


```python
df['rating'] = df['rating'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna(pd.Timestamp("1900-01-01"))
```

## . Standardizing `rating` and `duration`

- `rating`: Standardize values by ensuring consistent formatting (e.g., removing extra spaces)
- `duration`: Split into numeric value and unit (Minutes or Seasons) for easier analysis


```python
df['rating'] = df['rating'].str.strip()

df[['duration_value', 'duration_unit']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_value'] = pd.to_numeric(df['duration_value'], errors='coerce')
df['duration_unit'] = df['duration_unit'].str.capitalize()
```
